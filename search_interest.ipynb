```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
trends = pd.read_csv('ai_tools_comparison.csv')

# Inspect the data
trends.head()
```

```python
# Most consistent growth (fixed) 
trends['week'] = pd.to_datetime(trends['week'])
trends = trends.set_index('week')

pct_changes = trends[['chatgpt', 'gemini', 'microsoft_copilot']].pct_change()
std_devs = pct_changes.std()
most_consistent_tool = std_devs.idxmin()  # smallest std dev = most consistent

# Visualization + ChatGPT's biggest dip 
plt.figure(figsize=(12, 6))
for col in ['chatgpt', 'gemini', 'microsoft_copilot']:
    plt.plot(trends.index, trends[col], label=col, marker='o')

plt.title('AI Tools Search Interest Over Time')
plt.xlabel('Week')
plt.ylabel('Search Interest')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

chatgpt_changes = trends['chatgpt'].pct_change()
dip_index = chatgpt_changes.idxmin()
gpt_dip = dip_index.strftime('%B %Y')

#  Best month
trends['month_name'] = trends.index.strftime('%B')
trends['avg_interest'] = trends[['chatgpt', 'gemini', 'microsoft_copilot']].mean(axis=1)
monthly_avg = trends.groupby('month_name')['avg_interest'].mean()
best_month = monthly_avg.idxmax()

print("Most consistent tool:", most_consistent_tool)
print("ChatGPT biggest dip:", gpt_dip)
print("Best month:", best_month)
```
