```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.ccny.cuny.edu/registrar/fall"

response = requests.get(url)
html = response.text

print("Page downloaded successfully")
```

```python
soup = BeautifulSoup(html, "html.parser")

print(soup.title)
```

```python
rows = soup.find_all("tr")

print("Number of rows found:", len(rows))
```

```python
dates = []
dows = []
texts = []

for row in rows:
    cols = row.find_all("td")

    if len(cols) >= 3:
        date = cols[0].text.strip()
        dow = cols[1].text.strip()
        text = cols[2].text.strip()

        dates.append(date)
        dows.append(dow)
        texts.append(text)

print("Rows extracted:", len(dates))
```

```python
df = pd.DataFrame({
    "date": dates,
    "dow": dows,
    "text": texts
})

df.head()
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")

df.head()
```

```python
df = df.dropna(subset=["date"])

df.head()
```

```python
df = df.set_index("date")

df.head()
```

```python
df.to_csv("ccny_fall_2021_calendar.csv")

print("CSV file saved successfully")
```
