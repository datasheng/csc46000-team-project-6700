```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('games_details.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["PTS"])
```

```python
# convert minutes to numeric
df["MIN"] = pd.to_timedelta(df["MIN"], errors="coerce").dt.total_seconds() / 60
```

```python
num_columns=["PTS", "AST", "REB", "OREB", "DREB", "STL", "BLK", "TO", "PF", "MIN", "PLUS_MINUS", "FGM"]
cat_cols=["SEASON", "LEAGUE", "TEAM", "OPPONENT", "LOCATION", "W/L"]
df[num_columns].hist(figsize=(12, 10))
plt.suptitle("Histograms of Numerical Variables")
plt.tight_layout()
plt.show()
```

The right-skewed distrubution of points scored by players, asissts, and rebounds in the dataset suggests that most players tend to score fewer points, make fewer assists, and grab fewer rebounds in a game, while a smaller number of players (star players) achieve higher values in these categories.

```python
df[num_columns].describe()
```

Point scoring in the dataset shows a wide range, with a mean of 9.6 points per game and a maximum of 81 points (Kobe). I'm assuming that Wilt's 100 point game is not included in the dataset. This indicates that while players score relatively low, there are high scoring games as well.

Players also average 2.1 assists per game, with the max being 25.

Players average 4.05 rebounds per game, with a max of 18. Rebounds are typically dominated by centers.

The average field goals made is 3.58, which alligns with an average of 9.68 points scored per game. Thee max is 28 field goals made.

The average plus minus is -0.00048. Players tend to not play well most nights. The max is 57, which shows an exceptional game from a star player whose stats increased in the other numerical categories.




```python
df[num_columns].boxplot(figsize=(12, 10))
plt.xticks(rotation=45)
plt.title("Boxplot of Numerical Variables")
plt.xlabel("Numerical Variables")
plt.ylabel("Values")
plt.show()
```

Points scored by players, assists, and rebounds show a wide range of values with some extreme outliers.

There are games where star plays score significantly higher points, make more assists, and grab more rebounds than the average player, which is reflected in the presence of outliers in these variables.

```python
sns.pairplot(df[["PTS", "AST", "REB", "FGM","PLUS_MINUS"]])
```

The box plot shows the trend in basketball games:

The numerical variables have a positive relationship with each other, as players who score more points also tend to have more assists, rebounds, field goals made, and plus-minus.

Players with higher points scored often have a higher plus-minus rating (PLUS_MINUS).

```python
sns.countplot(x="START_POSITION", hue="START_POSITION", data=df, legend=False)
plt.title("Count of Player Positions")
plt.xlabel("Player Position")
plt.ylabel("Count")
plt.show()
```

The number of of players in each position

```python
sns.countplot(x="START_POSITION", hue="TEAM_ABBREVIATION", data=df)
plt.title("Count of Player Positions by Team")
plt.xlabel("Player Position")
plt.ylabel("Count")
plt.legend(title="Team", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Teams have different distributions of player positions, with some teams having more players in certain positions than others.

```python
sns.boxplot(x='START_POSITION', y='PTS', hue='START_POSITION', data=df, legend=False)
plt.title("Points Scored by Player Position")
plt.xlabel("Player Position")
plt.ylabel("Points Scored")
```

Since guards and forwards handle the ball more often and are more involved in scoring plays, they tend to score more points on average compared to centers, who typically focus more on rebounding and defense.

```python
sns.scatterplot(x='AST', y='PTS', hue='START_POSITION', data=df)
plt.title("Points vs Assists by Player Position")
plt.xlabel("Assists")
plt.ylabel("Points Scored")
plt.show()
```

Players who make more assists also tend to score more points, and this relationship is influenced by player position, with guards and forwards generally showing higher values in both assists and points compared to centers.
